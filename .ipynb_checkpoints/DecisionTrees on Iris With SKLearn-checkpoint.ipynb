```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
```

```python
iris = load_iris()
```

```python
X = iris.data[:, 2:]
y= iris.target
```

Let's train our model without preprocessing the data, as Decision Trees require no feature scaling or centering:

```python
tree_classifier = DecisionTreeClassifier(max_depth=2)
tree_classifier.fit(X, y)
```

```python
export_graphviz(tree_classifier,
               out_file="iris_tree.dot",
               feature_names=iris.feature_names[2:],
               class_names=iris.target_names,
               rounded=True,
               filled=True
               )
```

```python
with open ("iris_tree.dot")  as f:
    dot_graph = f.read()
    
display(graphviz.Source(dot_graph))
```

The Gini score indicates the purity of a given leaf (gini=0 if all instances it belongs to apply to one class). Note that SKLearn uses the CART algorithm for Decision Trees (binary trees only). When a leaf is pure, it cannot be split further (it has a straight and complete decision boundary).
